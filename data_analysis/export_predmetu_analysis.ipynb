```python
from collections import Counter
from pathlib import Path

import pandas as pd
import openpyxl

fav_predmety = pd.DataFrame(openpyxl.load_workbook(Path("../scrapers/data/exportPredmetu-FAV.xlsx")).active.values)
fel_predmety = pd.DataFrame(openpyxl.load_workbook(Path("../scrapers/data/exportPredmetu-FEL.xlsx")).active.values)
fst_predmety = pd.DataFrame(openpyxl.load_workbook(Path("../scrapers/data/exportPredmetu-FST.xlsx")).active.values)
```

```python
all_predmety = pd.concat([fav_predmety, fst_predmety, fav_predmety])
all_predmety
```

```python
def group_numbers(l: list):
    for i in range(len(l)):
        item = l[i]
        if item is None:
            continue

        try:
            int(item.split()[-1]) # check if course ends with a number
        except ValueError:
            continue

        l[i] = " ".join(item.split()[:-1]) # remove the number

def remove_nones(l: list):
    deleted = 0
    for i in range(len(l)):
        item = l[i-deleted]
        if item is None:
            del l[i-deleted]
            deleted += 1
```

```python
names = list(all_predmety[2])[2:]
remove_nones(names)
group_numbers(names)
```
